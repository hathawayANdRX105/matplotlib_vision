```python
import os, sys
import numpy as np
from pyecharts import options as opt


project_path = os.path.dirname(os.getcwd())
sys.path.append(project_path)

data_dir = r"{}\data\pyecharts_data".format(project_path)

city_file_name = "CityTS.7.3.xlsx"
city_data_path = r"{}\{}".format(data_dir, city_file_name)

print(data_dir, "\t", city_data_path)
```

```python
import pandas as pd
city_data = pd.read_excel(city_data_path, "CityTS0205")
city_data = city_data.fillna(0)
city_data
```

```python
sieve = city_data.loc[city_data['日期'] == r'2020/6/21'].copy()

def extract_name(name):
    if name.__contains__('省'):
        return name[:-1]
    else:
        return name

sieve['省市名称'] = sieve['省市名称'].map(extract_name)


sieve_data = pd.DataFrame(sieve, columns=['省市名称', '确诊人数'])

print(sieve_data)
```

```python
from pyecharts import options as opts
from pyecharts.charts import Map

provinces = sieve_data['省市名称'].values
comfirmed_count = sieve_data['确诊人数'].values

data = [(site, count) for site, count in zip(provinces, comfirmed_count)]
pieces = [
    {"min": 60000},
    {"min": 2000, "max": 60000},
    {"min": 1000, "max": 2000},
    {"min": 500, "max": 1000},
    {"min": 200, "max": 500,},
    {"min": 100, "max": 200,},
    {"min": 1, "max": 100, },
    {"value": 1}     
]

_map = (Map()
        .add("确诊人数", data, "china")
        .set_global_opts(
            visualmap_opts=opts.VisualMapOpts(max_=2000,
                                              pieces=pieces,
                                              is_piecewise=True, ),
            legend_opts=opts.LegendOpts(is_show=True,
                                        pos_left='left',
                                        pos_top='50%'),
            title_opts=opts.TitleOpts(
                title="2020/6/21 新冠肺炎确诊人数",
                subtitle="数据来源:中国卫健委统计发布",
                pos_right="center",
                pos_top="5%")))
_map.render_notebook()
```

```python
_dict = {}
for index, row in city_data.iterrows():
	row_date = row['日期']
	row_name = row['省市名称']
	row_count = row['确诊人数']

	item = _dict.get(row_name, {})

	if not item:
		new_item = {'first_date': row_date, 'cur_confirmed_count': row_count}

		_dict[row_name] = new_item
	elif item['cur_confirmed_count'] < row_count:
		item['last_date'] = row_date
		item['cur_confirmed_count'] = row_count


```

```python

import datetime as dt

# 西藏省 保持第一天 只感染1人, 补充最后一天为第一天
temp_item = _dict['西藏省']
temp_item['last_date'] = temp_item['first_date']

names = []
dates = []
for key, value in _dict.items():
    item = _dict[key]
    t1 = item['first_date']
    t2 = item['last_date']

    date1_str = t1.strftime('%Y-%m-%d')
    date2_str = t2.strftime('%Y-%m-%d')

    t1 = dt.datetime.strptime(date1_str, "%Y-%m-%d")
    t2 = dt.datetime.strptime(date2_str, "%Y-%m-%d")
    delta = t2 - t1

    if delta.days > 20:
        names.append(key)
        dates.append(delta.days)

print(names, "\n", dates)
print(len(names))
```

```python
from pyecharts import options as opts
from pyecharts.charts import Pie
import pandas as pd

pie_df = pd.DataFrame({'provinces': names, 'dates': dates})
pie_df.sort_values(by='dates', ascending=False, inplace=True)

provinces = pie_df['provinces'].values.tolist()
dates = pie_df['dates'].values.tolist()
# print("{}\n{}".format(provinces, dates))

color_series = [
    "#DC143C", "#DD1B42", "#DE2147", "#DF284D", "#E02F52", "#E13658",
    "#E23C5D", "#E34363", "#E44A69", "#E5506E", "#E65774", "#E75E79",
    "#E8657F", "#E96B84", "#EA728A", "#EB7990", "#EC7F95", "#ED869B",
    "#EE8DA0", "#EF94A6", "#F09AAB", "#F1A1B1", "#F2A8B7", "#F3AEBC",
    "#F4B5C2", "#F5BCC7", "#F6C3CD", "#F7C9D2", "#F8D0D8", "#F9D7DE",
    "#FADDE3", "#FBE4E9", "#FCEBEE", "#FDF2F4", "#FEF8F9",
    ] 

# opt settting
title_style_opt = opts.TextStyleOpts(font_size=25, color='#eb6238')
subtitle_textstyle_opt = opts.TextStyleOpts(font_size=50, color='#f8b500')

title_opt = opts.TitleOpts(
    title='多省区市确诊病例\n连续多日',
    subtitle='不间断',
    title_textstyle_opts=title_style_opt,
    subtitle_textstyle_opts=subtitle_textstyle_opt,
    pos_right='center',
    pos_left='center',
    pos_top='42%',
    pos_bottom='center',)


c = (Pie(init_opts=opts.InitOpts(width='1200px', height='800px'))
    .set_colors(color_series)
    .add("", 
         [list(z) for z in zip(provinces, dates)],
         radius=['30 % ', '100 % '],
         rosetype='area')
    .set_global_opts(title_opts=title_opt,
                     legend_opts=opts.LegendOpts(is_show=False),
                     toolbox_opts=opts.ToolboxOpts())
    
    .set_series_opts(label_opts=opts.LabelOpts(is_show=True,
                                               position='inside',
                                               font_size=8,
                                               formatter='{b}\n+{c}天',
                                               font_style='italic',
                                               font_weight='bold',
                                               font_family='Fangsong',
                                               color='#8A2BE2')))

c.render_notebook()
```
