```python
# 6-5 the attribute of axis
y_start = 0
y_end   = 1
axis_step  = 0.01
label_step = 0.2

data = np.arange(y_start, y_end, axis_step)  
ticks = np.arange(y_start, y_end, label_step)

plt.title('my lines example')

plt.xlabel('x')
plt.ylabel('y')

plt.xlim(y_start, y_end) # the start and end point same as y
plt.ylim(y_start, y_end)

plt.xticks(ticks)
plt.yticks(ticks)
plt.tick_params(labelsize=12)

plt.plot(data, data ** 2)
plt.plot(data, data ** 3)

plt.legend(['y=x^2', 'y=x^3'])

plt.show()
```

```python

# 6-6 tow picture


fig = plt.figure(figsize=(9,7), dpi=90)

ax1 = fig.add_subplot(1, 2, 1) # create first subplot
y_start = 0
y_end   = 1
axis_step  = 0.01
label_step = 0.2

data1 = np.arange(y_start, np.pi * 2, axis_step)  
ticks1 = np.arange(y_start, y_end, label_step)

plt.title('lines example')
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(y_start, y_end) # the start and end point same as y
plt.ylim(y_start, y_end)
plt.xticks(ticks1)
plt.yticks(ticks1)
plt.plot(data1, data1 ** 2)
plt.plot(data1, data1 ** 3)
plt.legend(['y=x^2', 'y=x^3'])


ax2 = fig.add_subplot(1, 2, 2) # create second subplot
x_start = 0
x_end   = np.pi * 2
y_start = -1
y_end   = 1
x_step = np.pi / 2
y_step = 0.5

data2 = data1

ticks2_x = np.arange(x_start, x_end, x_step)
ticks2_y = np.arange(y_start, y_end, y_step)

plt.title('sin-cos')
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(x_start, x_end) # the start and end point same as y
plt.ylim(y_start, y_end)
plt.xticks(ticks2_x)
plt.yticks(ticks2_y)
plt.plot(data2, np.sin(data2))
plt.plot(data2, np.cos(data2))
plt.legend(['sin', 'cos'])

plt.show()
```

```python
# 6-8

fig , ax = plt.subplots()

plt.rcParams['font.family'] = ['Microsoft YaHei UI'] 
plt.rcParams['axes.unicode_minus'] = False

def f(t):
    return np.cos(2 * np.pi * t)

x1 = np.arange(0, 0.4, 0.5)
x2 = np.arange(0, 4.0, 0.5)

plt.figure()

plt.subplot(2, 2, 1)
plt.title('子图1')
plt.plot(x1, f(x1), 'bo', x2, f(x2), 'k')

plt.subplot(2, 2, 2)
plt.title('子图2')
plt.plot(np.cos(f(x2)), 'r--')

plt.show()
```

```python
# 6-9
def getCumsum(size):
    return np.random.randn(size).cumsum()

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

ax.plot(getCumsum(30), color='k', linestyle='dashed', marker='o', label='one')
ax.plot(getCumsum(30), color='k', linestyle='dashed', marker='+', label='two')
ax.plot(getCumsum(30), color='k', linestyle='dashed', marker='v', label='three')
ax.legend(loc='best')
```

```python
# 6-12 fill
x = np.linspace(0, 1, 500) # different np.arrange(start, end ,step)
y = np.sin(3 * np.pi * x) * np.exp(-4 * x)

fig, ax = plt.subplots()

# range=x, upper_bound=0, lower_bound=y
plt.fill_between(x, 0, y, facecolor='green', alpha=0.3)

plt.plot(x, y)
```

```python
# 6-14
x = np.linspace(0, 1, 500) # generate 500 point between 0 and 1
y1 = np.sin(3 * np.pi * x) * np.exp(-4 * x)
y2 = y1 + 0.2

# omit creating the figture 
plt.plot(x, y1, 'b')
plt.plot(x, y2, 'r')
plt.fill_between(x, y1, y2, facecolor='green', alpha=0.3)
plt.show()
```

```python
# 6-15 latex
# $ latex formula $ inner
formula1 = r'$ \alpha \beta \pi \lambda \omega $'
formula2 = r'$ \sin(0) = \cos(\frac{ \pi }{2}) $'
formula3 = r'$ \lim_{x \rightarrow y} \frac{1}{x^3} $'
formula4 = r'$ \sqrt[4]{x} = \sqrt{y} $'
formula5 = r'$ \sum_{n=1}^{\infty}'r' \frac{ -e^{i \pi} }{ 2^n} $ !'

plt.xlim([1, 8])
plt.ylim([1, 3])

def plot_formula(x, y, formula, fontsize ):
    plt.text(x, y, formula, fontsize=fontsize)

plot_formula(2, 2, formula1, 15)
plot_formula(4, 2, formula2, 20)
plot_formula(2, 1.5, formula3, 25)
plot_formula(5, 1.5, formula4, 25)

plt.title(formula5, fontsize=20)
plt.show()
```
